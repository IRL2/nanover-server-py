```python
from nanover.app import OmniRunner
from nanover.openmm import OpenMMSimulation

simulation = OpenMMSimulation.from_xml_path("../ase/openmm_files/nanotube.xml")
simulation.load()

imd_runner = OmniRunner.with_basic_server(simulation, port=0, name="MOVEABLE RESTRAINTS")
imd_runner.load(0)
imd = imd_runner.app_server.imd
```

### Output atom indexes when interaction starts

```python
import ipywidgets as widgets
from nanover.imd import ParticleInteraction

def on_interaction_started(*, key: str, interaction: ParticleInteraction):
    with out:
        print(interaction.particles)

imd.interaction_started.add_callback(on_interaction_started)

out = widgets.Output()
out
```
