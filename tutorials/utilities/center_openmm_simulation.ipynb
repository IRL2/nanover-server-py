# Move system centroid to center of box

```python
INPUT_PATH = "../openmm/openmm_files/17-ala.xml"
OUTPUT_PATH = "17-ala-centered.xml"
```

Load OpenMM simulation:

```python
from nanover.openmm import OpenMMSimulation

simulation = OpenMMSimulation.from_xml_path(INPUT_PATH)
simulation.load()
```

Translate system centroid to box center:

```python
import numpy as np
from nanover.openmm.serializer import serialize_simulation

# extract positions from OpenMM
positions = np.array(simulation.simulation.context.getState(positions=True).getPositions())

# box center is half along each periodic box vector
center = np.sum(np.array(simulation.simulation.topology.getPeriodicBoxVectors()), axis=0) * .5
# offset by difference between current centroid and box center
offset = center - np.mean(positions, axis=0)
positions = positions + offset

# overwrite positions in OpenMM
simulation.simulation.context.setPositions(positions)
```

Save OpenMM simulation:

```python
with open(OUTPUT_PATH, "w") as outfile:
    outfile.write(serialize_simulation(simulation.simulation))
```
