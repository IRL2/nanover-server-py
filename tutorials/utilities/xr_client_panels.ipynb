# XR Client Panels

Examples of responding to generic user controller interaction.

## Setup runner & utilities

```python
from nanover.app import OmniRunner
from nanover.openmm import OpenMMSimulation

simulation = OpenMMSimulation.from_xml_path("../openmm/openmm_files/17-ala.xml")
simulation.load()

imd_runner = OmniRunner.with_basic_server(simulation, port=0, name="EXAMPLE: xr client cursors")
imd_runner.load(0)
```

```python
from nanover.jupyter import NanoverJupyterUtilities

utilities = NanoverJupyterUtilities.from_runner(imd_runner)
```

```python
def say_hello(intensity):
    utilities.notify_all(f"hell{"o" * int(intensity)}")

imd_runner.app_server.register_command("test/say-hello", say_hello)

utilities.panels.update_panel(
    "test",
    utilities.panels.header(label="Say Hello"),
    utilities.panels.slider(label="Intensity", range=(0, 5), variable="variable.hello-intensity"),
    utilities.panels.button(label="Execute", command="test/say-hello", arguments=dict(intensity={"variable": "variable.hello-intensity"})),
    label="Test Panel",
)
```

```python
print(imd_runner.app_server.copy_state()["panel.test"])
```
