# XR Client Panels

Example of showing a custom ui in the XR client. This is an experimental feature with client support only for a single panel under the key "test" which appears in the same menu as user commands.

## Setup runner & utilities

```python
from nanover.app import OmniRunner
from nanover.openmm import OpenMMSimulation

simulation = OpenMMSimulation.from_xml_path("../openmm/openmm_files/17-ala.xml")
simulation.load()

imd_runner = OmniRunner.with_basic_server(simulation, port=0, name="EXAMPLE: xr client cursors")
imd_runner.load(0)
```

```python
from nanover.jupyter import NanoverJupyterUtilities

utilities = NanoverJupyterUtilities.from_runner(imd_runner)
```

## Interface for running a command

Defining a command that sends a hello message to everyone, and an interface for modifying an input argument and triggering the command:

<video controls src="./assets/hello_panel.webm">

```python
def say_hello(intensity):
    utilities.notify_all(f"hell{"o" * int(intensity)}")

imd_runner.app_server.register_command("test/say-hello", say_hello)
```

```python
def hide_panel():
    utilities.panels.remove_panel("test")

def show_hello_panel():
    utilities.panels.update_panel(
        "test",
        utilities.panels.header(label="Config"),
        utilities.panels.slider(label="Intensity", range=(0, 5), variable="variable.hello-intensity", integer=True),
        utilities.panels.button(label="Say Hello", command="test/say-hello", arguments=dict(intensity={"variable": "variable.hello-intensity"})),
        utilities.panels.button(label="Close", command="panel/hide"),
        label="Hello editor",
    )

imd_runner.app_server.register_command("panel/hide", hide_panel)
imd_runner.app_server.register_command("user/show-hello", show_hello_panel, icon="👋", label="hello editor")
```
