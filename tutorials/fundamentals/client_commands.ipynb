# Registering commands as a client

## Introduction

Here we demonstrate the capability to implement commands on one client and register them on the server such that another client can call them. This works as long as the providing client remains available to provide the command.

## Setup bare server

```python
from nanover.app import OmniRunner

runner = OmniRunner.with_basic_server(port=0)
runner.print_basic_info()
```

## Connect first client and register command

```python
from nanover.websocket import NanoverImdClient

client_a = NanoverImdClient.from_runner(runner)
client_a.register_command("test/hello", lambda: {"hello": "world"})
```

## Connect second client and run command

```python
from nanover.websocket import NanoverImdClient

client_b = NanoverImdClient.from_runner(runner)
client_b.run_command_blocking("test/hello")
```
