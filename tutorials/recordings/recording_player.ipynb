# Serving frames of recording with seeking interface

## Introduction

This is a toy example showing how to use a jupyter notebook interface to serve individual frames of a recording.

## Load a recording and define a seek function

```python
from nanover.recording import PlaybackSimulation

simulation = PlaybackSimulation.from_path(name="Recording", path="nanotube-example-recording.nanover.zip")
simulation.load()

def seek(value):
    simulation.next_entry_index = 0
    simulation.advance_by_one_step()
    simulation.next_entry_index = value
    simulation.advance_by_one_step()

min_index = 0
max_index = len(simulation.entries) - 1

min_index, max_index
```

## Attach it to a server

```python
from nanover.app import OmniRunner

runner = OmniRunner.with_basic_server(simulation, name="Recording Playback Test", port=0)
runner.load(0)
```

## Attach seek function to a slider

```python
from ipywidgets import interact, widgets

interact(
    seek,
    value=widgets.IntSlider(
        default=min_index,
        description="Frame",
        min=min_index,
        max=max_index,
        value=min_index,
        layout={'width': 'auto'}
    ),
);
```

## Display in NGLView

```python
from nanover.nglview import NGLClient

client = NGLClient.from_runner(runner)
client.view
```
